In simple linear regression, the R-squared value is indeed always equal to the square of the Pearson correlation coefficient (r). However, this relationship becomes more nuanced in other scenarios. Let's break it down:

1. Simple Linear Regression:
For simple linear regression with one predictor variable, R-squared is always equal to the square of the Pearson correlation coefficient[1][2]. This can be expressed mathematically as:

$$R^2 = r^2$$

Where R^2 is the coefficient of determination and r is the Pearson correlation coefficient.

2. Multiple Linear Regression:
In multiple linear regression, the relationship still holds, but with a slight modification. R-squared is equal to the square of the correlation between the observed values and the predicted values from the regression model[4]. This can be written as:

$$R^2 = (Corr(y_{observed}, y_{predicted}))^2$$

3. Non-linear Relationships:
For non-linear relationships (e.g., exponential or logarithmic), the simple R^2 = r^2 relationship may not hold[1]. In these cases, the coefficient of determination (R^2) can still be calculated, but it may not have the same straightforward relationship with the Pearson correlation coefficient.

4. Interpretation:
- R-squared ranges from 0 to 1 and represents the proportion of variance in the dependent variable that is predictable from the independent variable(s)[2].
- The Pearson correlation coefficient (r) ranges from -1 to 1 and measures the strength and direction of the linear relationship between two variables[2].

5. Cautions:
- While R-squared and r are related, they have different interpretations. R-squared is about explained variance, while r is about the strength and direction of a linear relationship[5].
- These measures assume linear relationships. For non-linear relationships, other measures of association may be more appropriate[1].

In conclusion, while R-squared is always the square of r in simple linear regression, this relationship becomes more complex in multiple regression and non-linear models. It's important to consider the specific context and type of analysis when interpreting these measures.
