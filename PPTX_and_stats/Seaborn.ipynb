```python
import seaborn as sb
```

```python
df=sb.load_dataset("tips")
```

```python
df
```

```python
ax=sb.catplot(data=df, x="sex",y="tip",kind="swarm",size=4,alpha=.5)
sb.catplot(data=df,x="sex",y="tip",kind="box",fill=False,ax=ax)
```

```python
# g = sns.FacetGrid(data=indata, hue="Group", col ='Task',
#                   palette=["skyblue","salmon"])
# g.map(sns.boxplot, x="Emotion", y="Threshold", showmeans='True', meanline = 'True',
#                    height=6, aspect=0.8, boxprops={'facecolor':'None'}, edgecolor='gray')

# g.map(sns.swarmplot, x="Emotion", y="Threshold")

ax=sb.swarmplot(data=df, x="sex",y="tip",size=4,alpha=.5)
sb.boxplot(data=df,x="sex",y="tip",fill=False,ax=ax,showfliers=False,width=.25)
```

```python
g=sb.FacetGrid(data=df,col="sex",row="time")
order=list(set(df['day']))
g.map(sb.boxplot,x="day",y="tip",fill=False,showfliers=False,width=.25,data=df,order=order,color='black')
g.map(sb.swarmplot,y="tip",x="day",data=df,size=2,alpha=.75,order=order)
```

```python
g=sb.catplot(data=df, x="sex",y="tip",kind="box",col="time",row="smoker")
g.map(sb.swarmplot,data=df, x="sex",y="tip",size=4,alpha=.5)
```

```python
g=sb.FacetGrid(data=df,col="sex",row=None)
order=list(set(df['day']))
g.map(sb.boxplot,x="day",y="tip",fill=False,showfliers=False,width=.25,data=df,order=order,color='black')
g.map(sb.swarmplot,y="tip",x="day",data=df,size=2,alpha=.75,order=order)
```

```python
g=sb.FacetGrid(data=df,col=None,row=None)
order=list(set(df['day']))
g.map(sb.boxplot,x="day",y="tip",fill=False,showfliers=False,width=.25,data=df,order=order,color='black')
g.map(sb.swarmplot,y="tip",x="day",data=df,size=2,alpha=.75,order=order)
```
