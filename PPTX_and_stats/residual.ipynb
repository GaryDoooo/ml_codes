```python
def custom_plot(x, y, ax=None, **kwargs):
    if ax is None:
        ax = plt.gca()
    
    ax.plot(x, y, **kwargs)
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title('Custom Plot')
    
    return ax
import matplotlib.pyplot as plt
import numpy as np

# Create a figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Generate some example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Use the custom plot function for each subplot
custom_plot(x, y1, ax=ax1, color='blue', label='Sin')
custom_plot(x, y2, ax=ax2, color='red', label='Cos')

# Add legends
ax1.legend()
ax2.legend()

plt.tight_layout()
plt.show()
```
