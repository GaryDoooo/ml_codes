```python
def derivative_central(data):
    return [(data[i+1] - data[i-1])/2 for i in range(1, len(data)-1)]

def derivative_forward(data):
    return [data[i+1] - data[i] for i in range(len(data)-1)]
```

```python
x1, y1, x2, y2 =subs[1]
sub = a[y1:y2, x1:x2]
r = locate_cross(sub, cross_cnt=1)
x, y = r[0]
fig,axs=plt.subplots(2,2)
dys= [-40, -20, 20, 40]
cnt+=1
for i in range(4):
    dy=dys[i]
    l = [sub[y + dy][_] for _ in range(x - 20, x + 21)]
    x0 = x - 20 + l.index(max(l))
    l = [float(sub[y + dy][_]) for _ in range(x0 - 20, x0 + 20)]
    axs[i>>1][i&1].plot(derivative_forward(l))
plt.tight_layout()
# Add a main title to the entire figure
plt.suptitle(str(cnt), fontsize=16)

# Customize the outer frame
fig.patch.set_facecolor('lightgrey') 
plt.show()
```

```python
def compress_list(lst):
    return [x for i, x in enumerate(lst) if i == 0 or x != lst[i-1]]

# Example usage
original = [1, 1, 2, 3, 3, 3, 4, 4, 5, 1, 1]
compressed = compress_list(original)
print(compressed)  # Output: [1, 2, 3, 4, 5, 1]
```
