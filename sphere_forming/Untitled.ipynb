```python
import csv

with open('no_friction.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
```

```python
import matplotlib.pyplot as plt
```

```python
data=[[(float(cell)-1)*100 for cell in row] for row in data]
```

```python
plt.figure()
# plt.plot(data[50], 'blue')
plt.plot(data[0],'red')
plt.plot(data[20],'purple')
plt.plot(data[50],'green')
# plt.yscale("log") 
plt.xlabel('Segment #')
plt.ylabel('Stretch Ratio %')
plt.show()
```

```python
with open('with_friction.csv', newline='') as f:
    reader = csv.reader(f)
    data2 = list(reader)
```

```python
d=[[(float(cell)-1)*100 for cell in row] for row in data2]
```
