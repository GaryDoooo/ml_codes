```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def get_data(filename):
    with open(filename, newline='') as f:
        reader = csv.reader(f)
        data = list(reader)
    data=[[(float(cell)-1)*100 for cell in row] for row in data]
    s=data[0::2]
    t=data[1::2]
    return s,t

def plot_lines(data,idx_list=[0,25,50,75,-1]):
    plt.figure()
    for i in idx_list:
        plt.plot(data[i])

    plt.xlabel('Segment #')
    plt.ylabel('Stretch Ratio %')
    plt.show()

def polar_heatmap(data):    
    fig = plt.figure()
    ax = Axes3D(fig)

    rad = np.linspace(0, 1, len(data))
    azm = np.linspace(0, 2 * np.pi, 100)
    r, th = np.meshgrid(rad, azm)
    z = (r ** 2.0) / 4.0

    plt.subplot(projection="polar")

    plt.pcolormesh(th, r, z)

    plt.plot(azm, r, color='k', ls='none') 
    plt.grid()

    plt.show()
```

```python
#define MEMBERANCE_RADIUS 1
#define DOME_RADIUS 3
#define MOVE_THRESHOLD 1e-10
#define PRESSING_STEPS 1000
#define NUM_OF_SEGMENTS 100
#define FAN_SLICE_THETA 0.01
#define MASS_TIME_FACTOR 1e-4
#define YOUNGS_MODULUS 0.1
#define FRICTION_FACTOR 1
#define FRICTION_MOTION_TO_STATIC_RATIO 0.9
#define PRINT_EVERY_X_STEPS 10

s,t=get_data('down9.csv')
plot_lines(s)
```
