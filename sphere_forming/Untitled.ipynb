```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def get_data(filename):
    with open(filename, newline='') as f:
        reader = csv.reader(f)
        data = list(reader)
#     s=data[0::3]
    s=[[(float(cell)-1)*100 for cell in row] for row in data[0::3]]
#     t=data[1::3]
    t=[[(float(cell)-1)*100 for cell in row] for row in data[1::3]]
    d=[[float(cell) for cell in row] for row in data[2::3]]
    return s,t,d

def plot_lines(data,idx_list=[0,25,50,75,-1]):
    plt.figure()
    for i in idx_list:
        plt.plot(data[i])

    plt.xlabel('Segment #')
    plt.ylabel('Stretch Ratio %')
    plt.show()

def polar_heatmap(data,distance):    
#     fig = plt.figure()
#     ax = Axes3D(fig)

    rad = np.linspace(0, 1, len(data))
    azm = np.linspace(0, 2 * np.pi, 100)
    r, th = np.meshgrid(rad, azm)
    z = np.zeros(r.shape, dtype=float)
    for i in range(len(r)):
        for j in range(len(r[0])):
            r[i][j]=distance[j]
            z[i][j]=data[j]
            
    plt.style.use('classic')
    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True)
    c = ax.pcolormesh(th, r, z) 
    ax.set_thetamin(0)
    ax.set_thetamax(90)
    ax.set_rmax(max(d[-1]))
    fig.colorbar(c)
    plt.show()
            
#     plt.subplot(projection="polar")

#     plt.pcolormesh(th, r, z)
#     plt.colorbar()

#     plt.plot(azm, r, color='k', ls='none') 
#     plt.grid()
#     plt.yticks([0.2,0.4,0.6,0.8,1]," ") 

#     plt.show()
```

```python
DOME_RADIUS=3
from math import *

s,t,d=get_data('down9.csv')
s_last=[_/100+1 for _ in s[-1]]
t_last=[_/100+1 for _ in t[-1]]

rad = np.array(d[-1])
azm = np.linspace(0, 2 * np.pi, 1000)
r, th = np.meshgrid(rad, azm)
res=np.zeros(r.shape,dtype=float)

for ti in range(len(azm)):
    for ri in range(len(rad)):
        theta=th[ti][ri]
        dist=r[ti][ri]
        alpha=dist/DOME_RADIUS
        x=sin(theta)
        y=cos(theta)
        x_s=x*t_last[ri] # Stretched
        y_s=y*s_last[ri]
        y_p=y_s*cos(alpha) # projected to ground vertically
        new_theta=atan(x_s/y_p)
        res[ti][ri]=(new_theta-theta)/np.pi*180
        if(theta>np.pi/2):
            res[ti][ri]=0
        
plt.style.use('classic')
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
c = ax.pcolormesh(th, r, res) 

ax.set_thetamin(0)
ax.set_thetamax(90)
ax.set_rmax(max(d[-1]))

fig.colorbar(c)

plt.show()
```

```python
# QWP leakage
from numpy import linspace

thickness_gap=linspace(1,20,19)
leak=[]
for t in thickness_gap:
    p=np.pi/2*t/100;
    long=sqrt(2)*sin(np.pi/2-p/2)
    short=sqrt(2)*sin(p/2)
    leak.append(short*short/long/long)
    
plt.figure()
plt.plot(thickness_gap, leak, 'blue')

plt.xlabel('QWP stretched % thinner')
plt.ylabel('Ratio')
plt.grid('gray')
plt.yscale('log')
# plt.axis([400,800,0,ymax])

title='Light intensity ratio passed/reflected'
plt.title(title)
plt.show()
```

```python
# Misalignment leakage
from numpy import linspace

thickness_gap=linspace(0,5,20)
leak=[]
for t in thickness_gap:
    p=t/180*np.pi
    long=cos(p)*cos(p)
    short=sin(p)*sin(p)
    leak.append(short/long)
    
plt.figure()
plt.plot(thickness_gap, leak, 'blue')

plt.xlabel('Angle between polarization and reflection axis, degree')
plt.ylabel('Ratio')
plt.grid('gray')
plt.yscale('log')
# plt.axis([400,800,0,ymax])

title='Light intensity ratio passed/reflected'
plt.title(title)
plt.show()
```
