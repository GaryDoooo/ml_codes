```python
from alpha_vantage.timeseries import TimeSeries

ts = TimeSeries(key='EJ69MPM068NGTJ30', output_format='pandas',indexing_type='date')

data, meta_data = ts.get_daily_adjusted(symbol='qqq', outputsize='full')
print(data.head(),data.tail())
```

```python
av=data['2008-12-29':'2009-02-06'].copy()
av['adj']=av['5. adjusted close']/av['4. close']
av['ao']=av['adj']*av['1. open']
av['ah']=av['adj']*av['2. high']
av['al']=av['adj']*av['3. low']
av['ac']=av['5. adjusted close']
#pd.concat([df1['c'], df2['c']], axis=1, keys=['df1', 'df2'])
ticker='QQQ'
df=pd.concat([av['ao'],av['ah'],av['al'],av['ac'],av['6. volume']], axis=1, keys=[
    ticker+'_ao',ticker+'_ah',ticker+'_al',ticker+'_ac',ticker+'_vol'])
df
```

```python
pd.merge(df,s5,how='outer', left_index=True, right_index=True)
```
